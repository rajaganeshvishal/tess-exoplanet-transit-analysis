```python
import lightkurve as lk
import matplotlib.pyplot as plt

search = lk.search_lightcurve("HAT-P-7", mission="TESS")
lc = search[32].download(download_dir="R:\elden\images")
lc_clean = lc.remove_outliers()
lc_flat = lc_clean.flatten()
lc_flat.scatter()
lc_flat[2000:3000].scatter()
depth = 1 - lc_flat.flux.min()
plt.show()
print(f"Transit Depth: {depth:.4f}")
```

```python
depth = 1 - lc_flat.flux.min()
print(f"Transit Depth: {depth:.4f}")
import numpy as np
Rp_Rs = np.sqrt(depth)
print(f"P to S radius ratio (Rp/Rs): {Rp_Rs:.4f}")
print
period = 2.20473
lc_fold = lc_flat.fold(period=period)
ax = lc_fold.scatter()
ax.set_xlim(-0.9, -0.5)

plt.show()
```

```python
ax = lc_fold.scatter(alpha=0.3, s=5)

ax.set_xlim(-0.2, 0.2)
lc_fold.bin(time_bin_size=0.002).scatter(ax=ax, color='red', s=10)
plt.show()
```

```python
#fold first
t0 = lc_flat.time[lc_flat.flux.argmin()]
lc_fold = lc_flat.fold(period=period, epoch_time=t0)

lc_bin = lc_fold.bin(time_bin_size=0.002)

ax = lc_bin.scatter(color="red", s=25)
ax.set_xlim(-0.2, 0.2)
plt.show()
```
